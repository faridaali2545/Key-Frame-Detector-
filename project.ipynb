```python
import os
import zipfile
import cv2
import h5py
import numpy as np
```

```python
ZIP_PATH = r"C:\Users\dell\.vscode\deepProject\ydata-tvsum50-video.zip"
EXTRACT_DIR = r"C:\Users\dell\.vscode\deepProject\dataset\tvsum_videos"
FRAMES_DIR = r"C:\Users\dell\.vscode\deepProject\dataset\frames"
MAT_PATH = r"C:\Users\dell\.vscode\deepProject\ydata-tvsum50.mat"
```

```python
VIDEO_DIR = os.path.join(EXTRACT_DIR, "video")
```

```python
print("zip exists?", os.path.exists(ZIP_PATH))
os.makedirs(EXTRACT_DIR, exist_ok=True)

with zipfile.ZipFile(ZIP_PATH, "r") as z:
    z.extractall(EXTRACT_DIR)

print("Extraction complete!")
```

```python
os.makedirs(FRAMES_DIR, exist_ok=True)
```

```python
def extract_frames(video_path, output_dir, fps=1):
    os.makedirs(output_dir, exist_ok=True)
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    saved_count = 0
    video_fps = int(cap.get(cv2.CAP_PROP_FPS))
    success, frame = cap.read()
    while success:
        if frame_count % max(1, video_fps // fps) == 0:
            frame_file = os.path.join(output_dir, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(frame_file, frame)
            saved_count += 1
        frame_count += 1
        success, frame = cap.read()

    cap.release()
    print(f"Extracted {saved_count} frames out of {frame_count} frames from {os.path.basename(video_path)}")
```

```python
print("Looking for videos under EXTRACT_DIR:", EXTRACT_DIR)
# Search recursively for any .mp4 files inside the extracted archive
mp4_paths = []
for root, dirs, files in os.walk(EXTRACT_DIR):
    for f in files:
        if f.lower().endswith('.mp4'):
            mp4_paths.append(os.path.join(root, f))

if not mp4_paths:
    print("No .mp4 files found in the extracted archive.")
    print("Contents of EXTRACT_DIR:", os.listdir(EXTRACT_DIR))
else:
    # Use the directory that contains the mp4s (common parent if scattered)
    video_dir = os.path.commonpath([os.path.dirname(p) for p in mp4_paths])
    print("Using VIDEO_DIR:", video_dir)
    os.makedirs(FRAMES_DIR, exist_ok=True)
    for video_path in mp4_paths:
        video_file = os.path.basename(video_path)
        output_dir = os.path.join(FRAMES_DIR, os.path.splitext(video_file)[0])
        extract_frames(video_path, output_dir, fps=1)
```
